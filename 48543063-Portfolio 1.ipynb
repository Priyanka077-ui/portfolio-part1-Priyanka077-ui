## Analysis of a car sell record Dataset

We have been provided with a car sell record dataset. In this dataset, we have thounds of car sell records.

The structure of this dataset is represented in the header shown below.

name
year
selling_price
km_driven
fuel
seller_type
transmission
Owner

| name | year | selling_price | km_driven| fuel | seller_type | transmission | Owner |

    | ---- | ---- | ---- | ---- | ---- | ---- | ---- | ---- |   
    
#### Description of Fields

* __name__ - detail car type of the sold car
* __year__ - Year of the car when it was bought
* __selling_price__ - Price at which the car is being sold
* __km_driven__ - Number of Kilometres the car is driven
* __fuel__ - Fuel type of car (petrol / diesel / CNG / LPG / electric)
* __seller_type__ - Tells if a Seller is Individual or a Dealer
* __transmission__ - Gear transmission of the car (Automatic/Manual)
* __owner__ - Number of previous owners of the car.

There are four questions to explore with the data as shown below.

 ## Q1. Remove missing data

Please remove the records that match ANY of the below two conditions in the csv file:

__Display the DataFrame, counting number of Null values in each column, and print the length of the data__ before and after removing the missing data.  

```python
df = pd.read_csv("car_sell.csv")
df.head(10)
```

```python
print("Original DataFrame:")
(df)
```

```python
# Count the number of null values in each column before cleaning
null_counts_before = df.isnull().sum()
("\nNumber of null values in each column before cleaning:")
(null_counts_before)
```

```python
#printing length before cleaning the data
print(f"\nLength of the DataFrame: {len(df)}")
```

```python
#cleaning the year missing
df_year_cleaned = df.dropna(subset=['year'])
(df_year_cleaned.head(67))
```

```python
#displaying data after cleaning
df_year_cleaned.shape
(df_year_cleaned)
```

```python
#Counting number of null after cleaning the data of year
null_counts = df_year_cleaned.isnull().sum()

null_counts_df = null_counts.reset_index()
null_counts_df.columns = ['Column', 'Null Values']

null_counts_df
```

```python
#printing length after cleaning the data of year
print(f"\nLength of the DataFrame: {len(df_year_cleaned)}")
```

```python
#cleaning the fuel
df_fuel_cleaned = df[df['fuel'] != 'none'] 
(df_fuel_cleaned.head(25))
```

```python
df_fuel_cleaned.shape
(df_fuel_cleaned)
```

```python
#Merging the cleaned data set of both year and fuel to work for the further question
df_final_cleaned = df[(df['fuel'] != 'none') & df['year'].notna()]
(df_final_cleaned)
```

```python
#Counting number of null after cleaning the final data 
Final_null_counts = df_final_cleaned.isnull().sum()

Final_null_counts_df = Final_null_counts.reset_index()
Final_null_counts_df.columns = ['Column', 'Null Values']

Final_null_counts_df
```

```python
#printing length after cleaning the final data 
print(f"\nLength of the DataFrame: {len(df_final_cleaned)}")
```

```python
#Q2.3 descriptive statistics, including mean, std, max, and min of the number of selling_price grouped by different seller_type
grouped_sell_stats = df_final_cleaned.groupby('seller_type')['selling_price'].agg(count='count', mean='mean',std='std',min='min',max='max')
print(grouped_sell_stats.describe())
```

```python
#Q2.4 descriptive statistics, including mean, std, max, min of the number of years that sold by each owner type
grouped_owner_stats = df_final_cleaned.groupby('owner')['year'].agg(count='count', mean='mean',std='std',min='min',max='max')
print(grouped_owner_stats.describe())
```

```python
#importing the libraries that is needed for the further analysis
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Setting up the plotting area
plt.figure(figsize=(20, 10))

# 1. Boxplot for 'selling_price' by 'owner'
plt.subplot(2, 2, 1)
sns.boxplot(x='owner', y='selling_price', data=df_final_cleaned)
plt.title("Selling Price by Owner Type")
plt.xlabel("Owner Type")
plt.ylabel("Selling Price")
```

**Explanation**:

Cars with their first owner typically sell for more money than cars with a second or subsequent owner, according to boxplots. In general, you may see that the cars belonging to first-owners have a higher median selling price, although the range may be broader for other owner types. This implies that automobiles with fewer owners—especially those that are the first owners—are worth more in the market, while other owner types can exhibit more variation in selling prices, with certain categories having lower median prices and a larger range of values. There are a few possible reasons for this fluctuation, including condition, mileage, and maintenance history.


```python
plt.figure(figsize=(10,4))
for fuel_type in df_final_cleaned['fuel'].unique():
    subset = df_final_cleaned[df_final_cleaned['fuel'] == fuel_type]
    plt.hist(subset['selling_price'], bins=30, alpha=0.5, label=fuel_type)

plt.title('Histogram of Selling Prices by Fuel')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.legend(title='Fuel')
plt.show()
```

**Explaination**

The bulk of cars are sold for less than $1 million, and petrol and diesel cars are especially common in this price range. There is a considerable decline in the frequency of car sales as the selling price increases. Petrol cars have the most widespread fuel type distribution, with the highest frequency found in the lower price range. Diesel vehicles are slightly less common than gasoline-powered vehicles in the lower price range. Electric, LPG, and CNG vehicles are more limited in price ranges and have fewer frequencies. The increasing number of petrol and diesel vehicles in the lower price bracket might be attributed to their increased accessibility and appeal. The market for alternative fuel and electric vehicles appears to be in its infancy, based on the data.


```python
# 1. Boxplot for 'fuel' by 'km_driven'
plt.figure(figsize=(10, 6))
           

sns.boxplot(x='fuel', y='km_driven', data=df_final_cleaned)
plt.title("Fuel by Km Driven")
plt.xlabel("Fuel")
plt.ylabel("Km_driven")
```

**Explaination**

The plot shows that while CNG vehicles are driven less, maybe because they are used in urban settings, diesel cars are typically used more frequently and probably over longer distances.Diesel is chosen for larger mileage requirements and CNG for more cost-effective, shorter journeys. The variations in mileage reflect the common use cases and efficiencies of each fuel type.
Potential buyers or sellers may find it helpful to understand these patterns because increased mileage may signal more wear and tear, which could affect the vehicle's value and maintenance requirements.Whereas, the electric cars are almost not driven which we can see from the plot.



```python
# Scatter plot of 'owner' vs. 'year'
plt.figure(figsize=(10, 6))

sns.scatterplot(x='km_driven', y='year', data=df_final_cleaned, hue='km_driven')
plt.title("km_driven vs. Year")
plt.xlabel("km_driven")
plt.ylabel("Year")

plt.tight_layout()
plt.show()
```

**Explaination**

Understanding how km_driven varies with the year of manufacture is made easier with the use of the scatter plot. It is typical to see that newer cars may have comparatively lower mileage, while older cars typically have larger kilometres driven.
The graphic demonstrates that although cars accrue more km as they get older, there is a lot of variation. Even though they are newer, certain cars may have been driven more frequently, resulting in higher mileage. The data is based on a model year from 1995 to 2020 and color-coded points based on the distance driven. Due to their recent age, cars with km driven values below 200,000 km are often less than those with older models. The range of km driven is larger for older cars, many of which have higher mileage. Cars with abnormally high km_driven values are examples of outliers; these are probably very old or heavily used cars. Understanding the relationship between mileage and the age of the car and estimating its remaining life may be helped by this knowledge.


```python
plt.figure(figsize=(10,4))
for fuel_type in df_final_cleaned['fuel'].unique():
    subset = df_final_cleaned[df_final_cleaned['fuel'] == fuel_type]
    plt.hist(subset['transmission'], bins=30, alpha=0.5, label=fuel_type)

plt.title('Histogram of Transmission by Fuel')
plt.xlabel('Transmission')
plt.ylabel('Frequency')
plt.legend(title='Fuel')
plt.show()
```

**Explaination**

The frequency of cars with manual transmissions is noticeably higher than that of automatics, according to the histogram. The big bar for the manual transmission makes this clear.
This implies that the majority of the cars in the dataset are manual which is a sign of a trend in which the number of automated cars advertised or sold in this specific dataset is falling.
